```python
import csv
from itertools import islice
import base64
from IPython.display import Image, display

with open("employees_base64.csv", "r", newline="") as employeesFile:
    reader = csv.DictReader(employeesFile)
    for employee in islice(reader, 3):
        print(f"{employee['FirstName']} {employee['LastName']}")
        photo = employee['Photo']
        photo = base64.b64decode(photo)
        display(Image(data=photo))        
```
